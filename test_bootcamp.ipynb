```python
sns.heatmap(cm, 
			annot=True,
			fmt='g', 
            xticklabels= ['both', 'buy', 'other', 'sale'],
            yticklabels=['both', 'buy', 'other', 'sale']
			)
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```
